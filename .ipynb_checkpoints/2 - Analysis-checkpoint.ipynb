```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('claims_sample_data_cleared.csv')
```

```python
df.head(3)
```

# Check std of service share by month

The purpose of this block is to understand how share of concrete service changes over time 

```python
df1 = pd.merge(
    df.groupby(['MONTH']).agg({'PAID_AMOUNT': 'sum'}).reset_index(),    
    df.groupby(['MONTH','SERVICE_CATEGORY']).agg({'PAID_AMOUNT': 'sum'}).reset_index()\
    .pivot(index= 'MONTH',columns='SERVICE_CATEGORY', values='PAID_AMOUNT'),
    left_on = ['MONTH'],
    right_on = ['MONTH'])
```

```python
df1.columns
```

```python
for c in ['ASCServices', 'AncillaryFFS', 'ERServices',
       'InpatientServices', 'OutpatientServices', 'PCPEncounter', 'PCPFFS',
       'SNFServices', 'SpecialistFFS', 'SpecialistsFFS']:
    df1[c+'share'] = df1[c]/df1['PAID_AMOUNT']
```

```python
result1 = df1[['ASCServicesshare',
       'AncillaryFFSshare', 'ERServicesshare', 'InpatientServicesshare',
       'OutpatientServicesshare', 'PCPEncountershare', 'PCPFFSshare',
       'SNFServicesshare', 'SpecialistFFSshare', 'SpecialistsFFSshare']].describe(percentiles=[.05,0.95])
result1 = result1[result1.index != 'count']
result1
```

```python
for c in result1.columns:
    sns.factorplot(x=result1.index, y = c, data = result1)
```

# Check std of payer share by month

```python
df1 = pd.merge(
    df.groupby(['MONTH']).agg({'PAID_AMOUNT': 'sum'}).reset_index(),    
    df.groupby(['MONTH','PAYER']).agg({'PAID_AMOUNT': 'sum'}).reset_index().sort_values(by=['PAID_AMOUNT'])\
    .pivot(index= 'MONTH',columns='PAYER', values='PAID_AMOUNT'),
    left_on = ['MONTH'],
    right_on = ['MONTH'])
```

```python
df1.columns
```

```python
for c in ['Payer B', 'Payer CA', 'Payer CO', 'Payer F',
       'Payer H', 'Payer O', 'Payer S', 'Payer UL', 'Payer UN', 'Payer W']:
    df1[c+'share'] = df1[c]/df1['PAID_AMOUNT']
```

```python
result1 = df1[['Payer Bshare', 'Payer CAshare', 'Payer COshare', 'Payer Fshare',
       'Payer Hshare', 'Payer Oshare', 'Payer Sshare', 'Payer ULshare',
       'Payer UNshare', 'Payer Wshare']].describe(percentiles=[.05,0.95])
result1 = result1[result1.index != 'count']
result1
```

```python
for c in result1.columns:
    sns.factorplot(x=result1.index, y = c, data = result1)
```

## 84% of claims are covered by F, CA, H and O

```python
df1['FCAHOshare'] = df1['Payer CAshare'] + \
df1['Payer Fshare'] + \
df1['Payer Hshare'] + \
df1['Payer Oshare']
```

```python
df1['FCAHOshare'].describe(percentiles=[.05,0.95])
```

# Check correlation between Claim Specialties

## TOP N specialties

```python
topspec = df.groupby(['CLAIM_SPECIALTY_final'])\
.agg({'PAID_AMOUNT' : 'sum'})\
.reset_index()\
.sort_values(by = ['PAID_AMOUNT'], ascending = False)\
[0:20]['CLAIM_SPECIALTY_final'].values
```

```python
df2 = df[df['CLAIM_SPECIALTY_final'].isin(topspec)]\
.groupby(['CLAIM_SPECIALTY_final', 'MONTH'])\
.agg({'PAID_AMOUNT' : 'sum'})\
.reset_index()
```

```python
df2 = df2.pivot(index= 'MONTH',columns='CLAIM_SPECIALTY_final', values='PAID_AMOUNT').fillna(0).reset_index()
```

```python
df2.head(3)
```

```python
df2.columns
```

```python
df2 = df2.drop([ 'MONTH'], axis = 1)
```

```python
corr2 = df2.corr()
plt.figure(figsize=(30,10))
ax = sns.heatmap(
    corr2, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
```

```python
upper_corr_mat = corr2.where(
    np.triu(np.ones(corr2.shape), k=1).astype(np.bool))
unique_corr_pairs = upper_corr_mat.unstack().dropna()
sorted_mat = unique_corr_pairs.sort_values()
```

```python
sorted_mat[-10:]
```

```python
sorted_mat[:10]
```
